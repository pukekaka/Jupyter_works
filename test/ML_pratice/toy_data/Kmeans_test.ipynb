# 1. Random Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = -2 * np.random.rand(100,2)
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100, :] = X1

plt.scatter(X[:,0], X[:,1], s=20, c='b')
plt.show()
```

```python
plt.scatter(X1[:,0], X1[:,1], s=20, c='r')
plt.show()
```

```python
#plt.figure('x')
plt.scatter(X[:,0], X[:,1], s=20, c='b')
#plt.figure('x1')
plt.scatter(X1[:,0], X1[:,1], s=20, c='r')
plt.show()
```

## 1-1. K Means in Scikit Learn

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
```

```python
km.fit(X)
```

```python
km.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], s=20, c='b')
plt.scatter(2.02626236, 1.93371681, s=200, c='g', marker='s')
plt.scatter(-1.12398258, -1.04807709, s=200, c='r', marker='s')
plt.show()
```
