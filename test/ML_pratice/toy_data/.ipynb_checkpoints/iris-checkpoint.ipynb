# 2. Analysis

## 2-1 Decision Tree (ID3)

```python
#from sklearn.model_selection import train_test_split
#from sklearn import tree, linear_model
#from sklearn.cluster import KMeans
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_target)
```

```python
print('iris data: ', iris_data.shape, 'iris target: ', iris_target.shape)
```

```python
print('X_train: ', X_train.shape, 'y_train: ', y_train.shape)
```

```python
dt_clf_gini = tree.DecisionTreeClassifier()
```

```python
dt_clf_gini.fit(X_train, y_train)
```

```python
dt_clf_entropy = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
dt_clf_entropy.fit(X_train, y_train)
```

```python
print('[gini index] train accuracy: {}'.format(dt_clf_gini.score(X_train, y_train)))
print('[gini index] test accuracy: {}'.format(dt_clf_gini.score(X_test, y_test)))
print('[information gain] train accuracy: {}'.format(dt_clf_entropy.score(X_train, y_train)))
print('[information gain] test accuracy: {}'.format(dt_clf_entropy.score(X_test, y_test)))
```

```python
dt_clf_gini = tree.DecisionTreeClassifier(max_depth=4)
```

```python
dt_clf_gini.fit(X_train, y_train)
```

```python
print('[gini index] train accuracy: {}'.format(dt_clf_gini.score(X_train, y_train)))
print('[gini index] test accuracy: {}'.format(dt_clf_gini.score(X_test, y_test)))
```

### 2-1-1 Graphviz

```python
tree.export_graphviz(dt_clf_gini, out_file='iris_dt_clf_gini.dot', class_names=['0', '1', '2'], feature_names=iris_data.columns, impurity=False, filled=True)
```

### 2-1-2 Feature Importance

```python
print('feature importance: {}'.format(dt_clf_gini.feature_importances_))
```

```python
def plot_feature_importances_iris(model):
    n_features = iris_data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), iris_data.columns)
    plt.xlabel('feature importances')
    plt.ylabel('feature')
    plt.ylim(-1, n_features)
plot_feature_importances_iris(dt_clf_gini)
```
